```python
from boruta import BorutaPy
import pandas as pd
import numpy as np
from BorutaShap import BorutaShap, load_data
```

```python
file = pd.read_csv("JPM_feature.csv")
file = file.drop(["Unnamed: 0"],axis =1)
X = file.drop(["value_to_predict","Date","JPM_close","JPM_volume","SPY_close","SPY_volume"],axis =1)
y = pd.Series(file["value_to_predict"])
```

```python
Feature_Selector = BorutaShap(importance_measure='shap',classification=False)
Feature_Selector.fit(X=X, y=y, n_trials=12, sample=False,train_or_test = 'test', normalize=True,verbose=True)
```

```python
Feature_Selector.plot(which_features='all')
```

AMZN

```python
ticker = "AMZN"
```

```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(axis=1)
    return df[indices_to_keep].astype(np.float64)
```

```python
file = pd.read_csv(ticker+"_feature.csv")
file = file.drop(["Unnamed: 0","Date"],axis =1)
file = clean_dataset(file)
X = file.drop(["value_to_predict",ticker+"_close",ticker+"_volume","SPY_close","SPY_volume"],axis =1)
y = pd.Series(file["value_to_predict"])
```

```python
Feature_Selector = BorutaShap(importance_measure='shap',classification=False)
Feature_Selector.fit(X=X, y=y, n_trials=12, sample=False,train_or_test = 'test', normalize=True,verbose=True)
```

```python
Feature_Selector.plot(which_features='all')
```
